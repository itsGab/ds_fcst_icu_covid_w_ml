Com base nas colunas e seus dados, foi percebido que alguns dados são categoricos de sim ou não, se a pessoa possui um caracteristica ou não, e assim foi separado essas colunas, representadas pela variável `colunas_categoricas`.

```python
colunas_categoricas = ['PATIENT_VISIT_IDENTIFIER', 'AGE_ABOVE65', 
                       'AGE_PERCENTIL', 'GENDER', 'DISEASE GROUPING 1', 
                       'DISEASE GROUPING 2', 'DISEASE GROUPING 3', 
                       'DISEASE GROUPING 4', 'DISEASE GROUPING 5', 
                       'DISEASE GROUPING 6', 'HTN', 'IMMUNOCOMPROMISED', 
                       'OTHER', 'WINDOW', 'ICU']
```

Removendo as colunas categoricas de todas as colunas, foi possível separar as colunas com dados continuas, representada pela variável `colunas_continuas`

```python
colunas_continuas = [coluna for coluna in df_inicial.columns if coluna not in colunas_categoricas]
```

## *** Hipótese 1: 
As colunas com dados continuos estão dividas em, 
* 'MEDIAN', 
* 'MEAN', 
* 'MIN', 
* 'MAX', 
* 'DIFF', 
* 'DIFF_REL'

Acredito não haver grande variação entre elas que façam dessas colunas relevante, então irei deixar apenas a 
* 'MEAN'

### testar a hipóteses!!!!

Com uma análise superficial do data frame, foi notado que não a uma variação muito exprecisava entre as varias medidas da mesmo "exame" (a max, min, median, mean, diff), principalmente levando em conta que estaremos usando para o modelo de previsão somente a primeira janela (window)

```python
palavra = 'MEAN'
colunas_continuas_mean = [col for col in colunas_continuas if palavra in col]
```

aqui juntei as colunas categoricas a continuas com apenas o "mean" e representei a variavel com `colunas_restante`



```python
colunas_restante = colunas_categoricas + colunas_continuas_mean
```

defini o DataFrame atual de trabalho como `df_atual`

```python
df_atual = df_inicial[colunas_restante]
```
